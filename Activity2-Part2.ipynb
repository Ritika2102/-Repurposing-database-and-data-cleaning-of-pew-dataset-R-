```r
Frame <- read.csv(file="player.csv",header=TRUE,sep=",")
Frame
```

```r
library(sqldf)
data <- arrange(Frame,desc(overall_rating))
data1 <- head(data,10)
ques1 <- print("Which are the top 10 players by overall rating?")
ans1 <- sqldf("SELECT player_name,overall_rating FROM data1")
ans1
```
