```r
adata <- tbl(my_db,sql("SELECT home_team_api_id,away_team_api_id,home_team_goal,away_team_goal FROM MATCH"))
```

```r
adataF <- data.frame(adata)
adataF
```

We have almost got all the attributes except the goals so we do the following computations to get that particular attribute

```r
library(sqldf)
adatahome <- sqldf("SELECT home_team_api_id, sum(home_team_goal) as home_goal  FROM adataF GROUP BY home_team_api_id")
adataaway <- sqldf("SELECT away_team_api_id,sum(away_team_goal) as team_goal FROM adataF GROUP BY away_team_api_id")
awayhome <- sqldf("SELECT home_team_api_id as team_api_id, (home_goal+team_goal) as sum_goal FROM adatahome LEFT JOIN adataaway WHERE adatahome.home_team_api_id=adataaway.away_team_api_id")
```

```r
library(sqldf)
#teamfinale <- na.omit(finalteam)
#teamfinale
teamfinal <- sqldf("SELECT * FROM finalteam GROUP BY team_api_id" )
teamfinal

```

So finally combine the tables using merge command by commom attribute team_api_id

```r
teamfilefinale=merge(x=awayhome,y=teamfinal, by="team_api_id",all.x=TRUE)
teamfilefinale
write.csv(teamfilefinale,file="team.csv")
```
